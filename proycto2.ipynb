#Ejercicio No 3
- Crear una función que permita generar 3 variables cualitativas
ordinales a partir de las variables numéricas con los siguientes criterios:

★ Si es mayor a la “media + 1*desviación estándar” → Alto
★ Si está entre “media - 1*desviación estándar” y “media +
1*desviación estándar” → Medio
★ Si es menor a la “media - 1*desviación estándar” → Bajo
Nota: Mostrar con las nuevas variables generadas.
Elegir con libertad las variables a trabajar.
La función debe contar con su respectiva documentación.

```python
# seleccion de la lista de las columnas numericas del data ¿frame original
columnas = df5.select_dtypes(include=['float64', 'int']).columns
columnas
```

```python
def generar_variables_cualitativas_ordinales(data):
  '''
  Esta función genera 3 variables cualitativas ordinales a partir de las
  variables numéricas proporcionadas en la lista "variables".

    Parámetros
  ----------
      data: el data frame coin las collumnas numericas.

   Retorno
  ----------
      Un DataFrame  con las columas cualitativas agregadas.
  '''
  data_resul=data.copy()
  for i in data:
    media = data[i].mean()
    desviacion_estandar = data[i].std()

    data_resul[i + '_cualitativo'] = np.where(data[i] > media + desviacion_estandar, 'Alto',np.where(data[i] < media - desviacion_estandar, 'Bajo', 'Medio'))

  return data_resul

# calculo de las variables indicadas
df5resultado = generar_variables_cualitativas_ordinales(df5[columnas])

# Mostrar el dataframe con las nuevas variables
df5resultado
```

#Ejercicio No 5
- Crear una función en Python que devuelva un mapa de calor con la
correlación de todas las variables numéricas (incluye las generadas),
según país.

● La función debe contar con su respectiva documentación.

```python
import matplotlib.pyplot as plt

def mapacalor(df:pd.DataFrame, pais:str):
  '''
  Función para la creación de los  mapas de calor del pais indicado:

  Parámetros
  ----------
      df: el data frame .
      pais: el pais con el cual se generara el mapa de calor
  '''
  # Filtrar el DataFrame por el país indicado
  df_pais = df[df['Pais_Origen'] == pais]

  # Obtener las variables numéricas
  variables_numericas = df_pais.select_dtypes(include=['int64', 'float64']).columns

  # Calcular la grilla de correlación
  correlacion = df_pais[variables_numericas].corr()

  # Crear el mapa de calor
  plt.figure(figsize=(10, 8))
  #sns.heatmap(correlacion, annot=True, fmt=".2f", cmap="Blues")
  sns.heatmap(correlacion,
            xticklabels = correlacion.columns.values,
            yticklabels = correlacion.columns.values,
            annot = True,
            cmap = "Blues"  # viridis, Blues
            );
  plt.title(f"Mapa de calor de la correlación de las variables numéricas en el pais: {pais}")
  plt.show()

  return correlacion

```

```python
# Genera el mapa de calor
mapacalor(df=df5, pais='USA')
```

```python
# Genera el mapa de calor
mapacalor(df=df5, pais='Europe')
```

#Ejercicio No 6
- Crear una función en Python que devuelva las siguientes estadísticas
● Media
● Mediana
● Moda
● Trimedia
● Varianza
● Desviación estándar
● Percentiles: 25, 50 y 75
● Mínimo
● Máximo

```python
# Definición de la función de calculo
def estadisticas_calculada(df:pd.DataFrame, variable:str):
  '''
  Función para la creación de los  mapas de calor del pais indicado:

  Parámetros
  ----------
      df: el data frame .
      variable: la  variable a calcular

  Estadisticos
  ------------
  - Media
  - Mediana
  - Moda
  - Trimedia
  - Varianza
  - Desviación estándar
  - Percentiles: 25, 50 y 75
  - Mínimo
  - Máximo
  '''
  # se genera una Serie definido como Decimales
  estadisticas = pd.Series(dtype='float64')
  estadisticas['media'] = df[variable].mean()
  estadisticas['mediana'] = df[variable].median()
  estadisticas['moda'] = df[variable].mode().iloc[0]
  estadisticas['trimedia'] = df[variable].quantile(0.3) + 2 * df[variable].quantile(0.7)
  estadisticas['varianza'] = df[variable].var()
  estadisticas['desviacion_estandar'] = df[variable].std()
  estadisticas['percentil_25'] = df[variable].quantile(0.25)
  estadisticas['percentil_50'] = df[variable].quantile(0.5)
  estadisticas['percentil_75'] = df[variable].quantile(0.75)
  estadisticas['minimo'] = df[variable].min()
  estadisticas['maximo'] = df[variable].max()

  return estadisticas

#ejecucion de la funcion para obtener los resultados para la  variable CILINDRO
estadisticas_calculada(df5, 'Cilindros')
```
